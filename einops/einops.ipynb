```python
import einops
import numpy as np
```

```python
A = np.arange(12).reshape(3,4) # 3 by 4 matrix
B = np.arange(-4,4).reshape(4,2) # 4 by 2 matrix
```

# einsum

```python
print(einops.einsum(A, B, "i k,k j->i j"))
```

# rearrange

```python
print(einops.rearrange(A, "h w -> w h"))
```

# reduce

```python
X = np.arange(12).reshape(3, 4).astype(np.float32)
print(X)
```

```python
print(einops.reduce(X, "rows cols -> cols", reduction="mean"))
print()
print(einops.reduce(X, "rows cols -> rows", reduction="mean"))
print()
print(einops.reduce(X, "rows cols -> ", reduction="mean"))
# Keep dimensions
einops.reduce(X, "rows cols -> rows 1", reduction="mean")
```

# repeat

```python
X = np.arange(4)
print(X)
```

```python
print(einops.repeat(X, "w -> h w", h = 4))
print()
print(einops.repeat(X, "w -> w 3"))
```

```python
X = np.arange(4).reshape(2, 2)
print(X)
```

```python
print(einops.repeat(X, "h w -> h (2 w)"))
print()
print(einops.repeat(X, "h w -> h (w 2)"))
```
